```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import LogNorm
import numpy as np

from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

% matplotlib inline
```

## 1. Bias vs variance trade-off

### Definition of polynomial $f(x)$ and the function used to calculate the Error per datapoint:

```python
def f(x):
    y = -306 + 93.3*x + 177.65*x**2 - 52.4*x**3 - 22.35*x**4 + 3.8*x**5 + x**6
    return y

def get_Error(data_y, fitted_data_y):
    data_y_np = np.array(data_y)
    fitted_data_y_np = np.array(fitted_data_y)
    
    avg_error = np.mean((data_y_np-fitted_data_y_np)**2)
    
    return avg_error
```

### Examining the number of data points vs. the Error:
#### The errors of a polynomial fit of (correct) degree 6 are calculated for different numbers of data points. For each number of data points the errors are averaged over 100 data-sets.

```python
num_datapoints = np.arange(10, 101, 1)
poly = PolynomialFeatures(degree = 6)
E_in_ = []
E_out_ = []
for i in num_datapoints: # loop calculates error for different numbers of data points
    E_in_i = []
    E_out_i = []
    for j in range(100): # loop calculates errors for 100 data sets for each number of data points
    
        data_x = np.linspace(-5.5, 3.5, i)

        noise_train = np.random.normal(loc = 0, scale = 2, size = i)
        train_data_y = f(data_x) + noise_train

        noise_test = np.random.normal(loc = 0, scale = 2, size = i)
        test_data_y = f(data_x) + noise_test


        data_X = poly.fit_transform(data_x[:, np.newaxis])
        clf = linear_model.LinearRegression()
        clf.fit(data_X, train_data_y)
        fitted_data_y = clf.predict(data_X)


        E_in_i.append(get_Error(train_data_y, fitted_data_y))
        E_out_i.append(get_Error(test_data_y, fitted_data_y))
    
    E_in = np.mean(E_in_i)
    E_out = np.mean(E_out_i)
    
    E_in_.append(E_in)
    E_out_.append(E_out)
```

#### Plotting the Errors of the fit with the training-data ($E_{in}$) and the test-data ($E_{out}$):
#### The decreasing behaviour of $E_{out}$ and the increasing behaviour of $E_{in}$ and their convergence towards a bias value of around 4 can be clearly seen in the plot.

```python
plt.figure(figsize = (10, 8))

plt.plot(num_datapoints, E_in_, ls = "None", marker = ".", color = "b", label = "Error with training-data $E_{in}$")
plt.plot(num_datapoints, E_out_, ls = "None", marker = ".", color = "g", label = "Error with test-data $E_{out}$")
plt.grid()
plt.legend()

plt.title("Error $E$ of fitted data over number of fitted data points $N$")
plt.xlabel("Number of data points $N$")
plt.ylabel("Error $E$ of fit as average squared\ndeviation per data point")
```

### Examining the model complexity vs. the Error:

#### The Errors with the test-data ($E_{out}$) are calculated for fits with polynomials of different degree.

```python
degrees = np.arange(1, 50, 1)



data_x = np.linspace(-5.5, 3.5, 100)

noise_train = np.random.normal(loc = 0, scale = 2, size = 100)
train_data_y = f(data_x) + noise_train

noise_test = np.random.normal(loc = 0, scale = 2, size = 100)
test_data_y = f(data_x) + noise_test

E_out_ = []
for i in degrees: # loop calculates error for different degrees of polynomials
    poly = PolynomialFeatures(degree = i)
    
    
    data_X = poly.fit_transform(data_x[:, np.newaxis])
    clf = linear_model.LinearRegression()
    clf.fit(data_X, train_data_y)
    fitted_data_y = clf.predict(data_X)


    E_out = get_Error(test_data_y, fitted_data_y)
    
    E_out_.append(E_out)
```

#### Plotting the Errors ($E_{out}$) over the degree of the fitted polynomial:
#### In can be seen that for low degrees of the polynomials the Error ist very high. This is because the polynomials here are too simpel to fit the full data-set and therefore in this region there is a low variance and a high bias. For large degrees the error is also high because here overfitting takes place and while the bias is low the variance increases strongly. Between these two regions the degrees with ideal complexity result in a comparably low error.

```python
plt.figure(figsize = (10, 8))

plt.plot(degrees, E_out_, ls = ":", marker = ".", ms = 10)
plt.grid()

plt.xlabel("Model-Complexity in degree of the fitted polynomial")
plt.ylabel("Average Squared Error $E_{out}$ per data point")
plt.title("Error of polynomial fits with test-data $E_{out}$ over the polynomial complexity (degree)")
```

## 2. Gradient descent - Adam

### Defining all necessary functions:

```python
def beales_function(x,y):
    f=np.square(1.5-x+x*y)+np.square(2.25-x+x*y*y)+np.square(2.625-x+x*y**3)
    return f

def grad_beales_function(params):
    x=params[0]
    y=params[1]
    grad_x=2*(1.5-x+x*y)*(-1+y)+2*(2.25-x+x*y**2)*(-1+y**2)+2*(2.625-x+x*y**3)*(-1+y**3)
    grad_y=2*(1.5-x+x*y)*x+4*(2.25-x+x*y**2)*x*y+6*(2.625-x+x*y**3)*x*y**2
    return [grad_x,grad_y]

def contour_beales_function():
    #plot beales function
    x, y = np.meshgrid(np.arange(-10.5, 10.5, 0.01), np.arange(-10.5, 4.5, 0.01))
    fig, ax = plt.subplots(figsize=(10, 10))
    z=beales_function(x,y)
    ax.contour(x, y, z, levels=np.logspace(0, 5, 45), norm=LogNorm(), cmap="RdYlBu_r")
    ax.plot(3,0.5, 'r*', markersize=18)

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')

    ax.set_xlim((-10.5, 10.5))
    ax.set_ylim((-4.5, 4.5))
    
    return fig,ax

def overlay_trajectory_contour_M(ax,trajectory, label,color='k',lw=2):
    xs=trajectory[:,0]
    ys=trajectory[:,1]
    ax.plot(xs,ys, color, label=label,lw=lw)
    ax.plot(xs[-1],ys[-1],color+'>', markersize=14)
    return ax;

def Adam(grad, init, n_epochs=5000, eta=10**-3, beta_1=0.9, beta_2 = 0.99, epsilon=10**-8, noise_strength=0):
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init
    s = 0
    m = 0
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        
        g = np.array(grad(params))+noise
        m = beta_1*m + (1 - beta_1)*g
        s = beta_2*s + (1 - beta_2)*g*g
        
        m_ = np.divide(m,(1 - beta_1**(j+1)))
        s_ = np.divide(s,(1 - beta_2**(j+1)))
        v = eta*np.divide(m_,(np.sqrt(s_) + epsilon))
        
        params = params - v
        param_traj[j+1,] = params
    return param_traj
```

### Behaviour of Adam Optimizer for two different starting points and two different learning rates:

```python
Nsteps = 10**4

init_1 = np.array([-2,3])
init_2 = np.array([5, -4])

lr_1 = 10**-3
lr_2 = 1


Adam_traj_i1_lr1 = Adam(grad_beales_function, init_1, Nsteps, eta = lr_1)
Adam_traj_i1_lr2 = Adam(grad_beales_function, init_1, Nsteps, eta = lr_2)
Adam_traj_i2_lr1 = Adam(grad_beales_function, init_2, Nsteps, eta = lr_1)
Adam_traj_i2_lr2 = Adam(grad_beales_function, init_2, Nsteps, eta = lr_2)
```

```python
fig1, ax1=contour_beales_function()
overlay_trajectory_contour_M(ax1, Adam_traj_i1_lr1, color = 'k', label = 'Adam trajectory with initial point {} and learing rate {}'.format(init_1, lr_1))

ax1.legend()
```

```python
fig1, ax1=contour_beales_function()
overlay_trajectory_contour_M(ax1, Adam_traj_i1_lr2, color = 'b', label = 'Adam trajectory with initial point {} and learing rate {}'.format(init_1, lr_2))

ax1.legend()
```

```python
fig1, ax1=contour_beales_function()
overlay_trajectory_contour_M(ax1, Adam_traj_i2_lr1, color = 'r', label = 'Adam trajectory with initial point {} and learing rate {}'.format(init_2, lr_1))

ax1.legend()
```

```python
fig1, ax1=contour_beales_function()
overlay_trajectory_contour_M(ax1, Adam_traj_i2_lr2, color = 'g', label = 'Adam trajectory with initial point {} and learing rate {}'.format(init_2, lr_2))

ax1.legend()
```
