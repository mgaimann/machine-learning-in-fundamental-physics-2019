```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpp
from sklearn import svm
import tensorflow as tf
import random
import palettable
from matplotlib.colors import LogNorm
```

```python
### Recycled functions from Exercise Sheet 1

# generate noisy data
def generate_noisydata(xstart, xend, bins, polycoeff_true, f):
    # initialize x and y arrays
    bins = int(bins)
    xdata = np.linspace(xstart, xend, num=bins+1)
    ydata = np.zeros(bins+1)

    # modify ydata by adding Gaussian random variables
    gaussians = np.zeros(len(xdata))
    ind = 0
    for g in gaussians:
        gaussians[ind] = random.gauss(0,150)
        ind += 1
    ydata = f(xdata, polycoeff_true) + gaussians    
    
    return xdata, ydata


# fit polynomials
def fitpolys(polys, xstart, xend, xdata, ydata, fit_xdata, bins, fig_ind=1, plot=True):
    ''' xdata, ydata: data to be fitted
    xstart, xend: specifies the boundaries of the fitting interval
    fit_xdata: xdata for fit
    bins: number of data points in specified array
    fig_ind: figure index for mpl's figure '''
    fitcoeffs = []
    fitcovariances = []

    # generate fit coefficients for all specified polynomial degrees
    # using numpy's polyfit
    ind = 0
    for p in polys:
        fitcoeff, fitcovariance = np.polyfit(xdata, ydata, deg=p, cov=True)
        fitcoeffs.append(fitcoeff)
        fitcovariances.append(fitcovariance)
        ind += 1

    # initialize overall fit data arrays and generate fit data:
    # take coefficient array for every polynomial
    if plot == True:
        plt.figure(fig_ind)
        fit_ydatas = []
        
        ind = 0
        for coeff in fitcoeffs:
            # initialize ydata array for current polynomial
            poly_degree = 0
            fit_ydata = np.zeros(len(fit_xdata))
            coeff = coeff[::-1] # start with coeff from lowest polynomial

            # create ydata for the current polynomial:
            # the fit is given by sum[ c_i * x^(p_i) ]
            for c in coeff:
                fit_ydata += c * (fit_xdata ** poly_degree)
                poly_degree += 1 

            # append generated data to overall fit array and plot
            fit_ydatas.append(fit_ydata)
            label = "%d" % polys[ind]
            plt.plot(fit_xdata, fit_ydata, label=label, c="r")
            ind += 1

        # plot data points
        plt.scatter(xdata, ydata)
        plt.legend(title="Polyn. Deg.")
        plt.xlabel("x")
        plt.ylabel("f(x)")
    
    return fitcoeffs, fitcovariances
```

```python
### A1 #### draw noisy data from deg 3 polynomial
# for different number of data points
# compute Eout and plot them


# set interval
xstart = -5
xend = 8
# set polynomial coefficients
polycoeff_true = [4.3, 0, -3.1, 7.2]
# set arbitrary xdata for plotting fit
fit_xdata = np.linspace(xstart, xend, 1000)


# define function, takes array with polynomial coefficients
# highest polynomials first (as in polyfit)
def f(x, polycoeff_true):
    polycoeff_true = polycoeff_true[::-1] # start with coeff from lowest polynomial
    y = np.zeros(len(x))
    poly_degree = 0
    for c in polycoeff_true:
        y += c * x**poly_degree
        poly_degree += 1
    return y


# compute errors E_out using true function vs. model function
# takes a set of true coeffiecients and a set of model coefficients
def calc_Eout(f, set_xdata, polycoeff_true, fitcoeff):    
    Eout = np.sum(np.absolute((f(set_xdata, polycoeff_true) - f(set_xdata, fitcoeff)))**2)
    return Eout
```

```python

### A1 - a ###
# draw noisy data from deg 3 polynomial
# for different number of data points
# compute Eout and plot them

# generate noisy data with different number of data points
bins_arr = np.logspace(1,5, 10) # number of data points considered
polys = [3] 
Eout_bins = []

# iterate through all bins
fig_ind = 0
for bins in bins_arr:
    # generate current reduced data set
    xdata, ydata = generate_noisydata(xstart, xend, bins, polycoeff_true, f)
    # fit the polynomial and plot
    fitcoeffs, fitcovariances = fitpolys(polys, xstart, xend, xdata, ydata, fit_xdata, bins, fig_ind = fig_ind, plot=True)
    Eout = calc_Eout(f, fit_xdata, polycoeff_true, fitcoeffs[0])
    Eout_bins.append(Eout)
    fig_ind += 1

# plot the errors for all bins
# we normalize the error by the number of data points
plt.figure()
plt.loglog(bins_arr, Eout_bins/bins_arr) 
plt.title("No. of Data Points vs. Error")
plt.xlabel("N")
plt.ylabel("Normalized Error")   
```

A1 - a - Discussion:
 - For a small number of data points the error rate is the highest. The statistics do not suffice for the noise to vanish.
 - The normalized error drops roughly as a power law with increasing number of data
 - For a vast amount of data, the noise is effectively averaged out and the error becomes very small

```python

### A1 - b ###
# NOW: draw noisy data from VARIOUS polynomials
# for SAME different number of data points
# compute Eout and plot

# generate noisy data with different number of data points
bins_arr = [11] # number of data points considered
polys = np.linspace(0,10,11)
Eout_polys = []

bins = bins_arr[0]
# generate current reduced data set
xdata, ydata = generate_noisydata(xstart, xend, bins, polycoeff_true, f)

# iterate through all polynomials
fig_ind = 0
for p in polys:
    # fit the polynomial and plot
    fitcoeffs, fitcovariances = fitpolys([p], xstart, xend, xdata, ydata, fit_xdata, bins, fig_ind = fig_ind, plot=True)
    Eout = calc_Eout(f, fit_xdata, polycoeff_true, fitcoeffs[0])
    Eout_polys.append(Eout)
    fig_ind += 1

# plot the errors for all polys
plt.figure()
plt.semilogy(polys, Eout_polys)
plt.title("Model Complexity vs. Error")
plt.xlabel("Model Complexity")
plt.ylabel("Error")   
```

A1 - b - Discussion:
 - Assume the underlying true polynomial has degree p (here: p=3).
 - For p_model < p we observe a significantly higher error (underfitting)
 - For p_model = p the error is lowest
 - For p_model > p we observe a slowly rising error which is due to fitting noisy contributions to the data (overfitting). Here, Bias is high, but Variance is low (we fit every data point, no residual error)!
 


```python
### A2 ###
# Adam Gradient Descent

# functions from lecture:
def beales_function(x,y):
    f=np.square(1.5-x+x*y)+np.square(2.25-x+x*y*y)+np.square(2.625-x+x*y**3)
    return f

def grad_beales_function(params):
    x=params[0]
    y=params[1]
    grad_x=2*(1.5-x+x*y)*(-1+y)+2*(2.25-x+x*y**2)*(-1+y**2)+2*(2.625-x+x*y**3)*(-1+y**3)
    grad_y=2*(1.5-x+x*y)*x+4*(2.25-x+x*y**2)*x*y+6*(2.625-x+x*y**3)*x*y**2
    return [grad_x,grad_y]

def contour_beales_function():
    #plot beales function
    x, y = np.meshgrid(np.arange(-10.5, 10.5, 0.01), np.arange(-10.5, 4.5, 0.01))
    fig, ax = plt.subplots(figsize=(10, 10))
    z=beales_function(x,y)
    ax.contour(x, y, z, levels=np.logspace(0, 5, 45), norm=LogNorm(), cmap="RdYlBu_r")
    ax.plot(3,0.5, 'r*', markersize=18)

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')

    ax.set_xlim((-10.5, 10.5))
    ax.set_ylim((-4.5, 4.5))
    
    return fig,ax

def overlay_trajectory_contour_M(ax,trajectory, label,color='k',lw=2):
    xs=trajectory[:,0]
    ys=trajectory[:,1]
    ax.plot(xs,ys, color, label=label,lw=lw)
    ax.plot(xs[-1],ys[-1],color+'>', markersize=14)
    return ax;

# modified version that plots only the trajectory with a color hexcode
def overlay_trajectory_contour_COL(ax,trajectory,color='k',lw=2):
    xs=trajectory[:,0]
    ys=trajectory[:,1]
    ax.plot(xs,ys, color,lw=lw)
    return ax;

# ADAM gradient descent:

def ADAM(init, grad, epochs, learningrate=1e-3, beta=np.array([0.9,0.99]), eps=1e-8):
    # initialize
    params = np.array(init)
    param_traj=np.zeros([epochs+1,2])
    param_traj[0,]=init
    m = 0
    v = 0
    
    # iterate
    for j in range(epochs):
        m = beta[0] * m + (1-beta[0]) * np.asarray(grad(params)) # exponentially decaying average of past gradients
        v = beta[1] * v + (1-beta[1]) * np.asarray(grad(params))**2 # average of past squared gradients
        m_hat = m / (1-beta[0]**(j+1)) # bias-corrected rates
        v_hat = v / (1-beta[1]**(j+1))
        params = params - learningrate / (np.sqrt(v_hat)+eps) * m_hat
        param_traj[j+1,] = params
        
    return param_traj
        
        
```

```python
# initialize parameters
epochs = 10**4
inits = [np.array([-2,3]), np.array([5,-4])]
learningrates = [1e-3, 1]

# check behaviour for different starting points and learning rates
for init in inits:
    for learningrate in learningrates:

        # plot contour lines
        fig1,ax1 =contour_beales_function()
        plt.title("Initial Cond.: %s / Learn.rate: %s" % (str(init), str(learningrate))) 
        
        # generate a trajectory
        traj = ADAM(init, grad_beales_function, epochs, learningrate=learningrate)
        overlay_trajectory_contour_M(ax1,traj, 'GDM','k')

        
```

A2 - a - Discussion:
 - For sufficiently low learning rates the respective closest minima are approached correctly.
 - For too high learning rates the algorithm overshoots, which leads to entering a different minimum or continuous switching between two minima
 - The choice of a suitable learning rate is crucial for the success of the gradient descent

```python
# scan over grid of starting points
xinits = np.linspace(-8,8,5)
yinits = np.linspace(-4,4,5)

# initialize
epochs = 10**4
learningrate = 1e-3
fig1,ax =contour_beales_function()
        
for x in xinits:
    for y in yinits:
        init = [x,y]

        # generate trajectory
        traj = ADAM(init, grad_beales_function, epochs, learningrate=learningrate)

        # extract start and end points and plot them with same color
        xs=traj[:,0]
        ys=traj[:,1]
        color = next(ax._get_lines.prop_cycler)['color']
        
        ax.plot(xs[-1],ys[-1],color, marker="s", markersize=14)
        overlay_trajectory_contour_COL(ax,traj, color=color)
        ax.plot(xs[0],ys[0],color, marker=">", markersize=14)
        
        #print("Computed %s" % str(init))
        
```

Obtain probability of field configuration from partition function:
 - The partition function is given in the form:
 - Z = sum_sigma exp(-beta * H(sigma))
 - where we sum over all configurations sigma. H(sigma) is the associated energy of a certain configuration and beta = 1/(kb * T), kb being the Boltzmann constant
 - The probability of a certain field configuration is then given by the Boltzmann distribution:
 - P(sigma) = 1/Z * exp(-beta * H(sigma))
 
 
Algorithms that allow drawing from an arbitrary probability distribution:
 - Inverse transform sampling
 - Rejection sampling
 - Composition of probability distributions
 
 
Ensure that generated samples are independently drawn from the probability distribution:
 - Apply the Kolmogorov-Smirnov test or the Cucconi test to two samples that were drawn from the distribution. If the test rejects, the samples were not drawn independently.
