```python
# Imports useful libraries

import numpy as np
import matplotlib.pyplot as plt
import tensorflow

from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_moons, make_circles, make_blobs
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from pandas import DataFrame


from tensorflow.keras.models import Model, Sequential # imports network type
from tensorflow.keras.layers import Dense, Input # imports our layers want to use

from tensorflow.python.keras.losses import binary_crossentropy, categorical_crossentropy # loss function
from tensorflow.python.keras.optimizers import Adam, SGD # optimizers
```

In the following we generate labelled data using the suggested libraries and we then implement the requested clustering algorithms. We start with blobs and then switch to circles and moons.

```python
# Generate sample data (BLOBS)
X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=4)

# Plot sample data with true classification
df = DataFrame(dict(x=X[:, 0], y=X[:, 1], label=y))
colors = {0: 'red', 1: 'blue', 2: 'green'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', color=colors[key])
plt.title('True Classification')
plt.show()
```

```python
for linkage in ('ward', 'complete', 'average', 'single'):
    model = AgglomerativeClustering(affinity='euclidean', compute_full_tree='auto', connectivity=None, linkage=linkage, n_clusters=3, pooling_func='deprecated')
    model.fit(X)
    y_predicted = model.fit_predict(X)

    # Plot data with predicted classification
    df_predicted = DataFrame(dict(x=X[:, 0], y=X[:, 1], label=y_predicted))
    grouped_predicted = df_predicted.groupby('label')
    colors_predicted = {0: 'red', 1: 'green', 2: 'blue'}
    fig_predicted, ax_predicted = plt.subplots()
    for key, group in grouped_predicted:
        group.plot(ax=ax_predicted, kind='scatter', x='x', y='y', color=colors_predicted[key])
    performance = str(np.floor(100 - np.linalg.norm(y-y_predicted)))
    plt.title('Classification ' + '"' + linkage + '"')
    plt.show()
    print('Accuracy ' + performance + '%')

# Comparison with k-Means
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
y_kmeans = kmeans.labels_
performance_kmeans = str(np.floor(100 - np.linalg.norm(y-y_kmeans)))
print('KMeans Accuracy ' + performance_kmeans + '%')
```

## Ex. 2 AutoEncoder

### 2.1 Generate 2-d images (1-channel) showing polynomials up to a maximum degree (e.g. 40x40) in two variables.

Here we create 40x40 images of two variable, given maximum degree polynomials. For semplicity we exclude cross terms in the two variables $x$ and $y$ as for example $xy, x^2y$, etc.

```python
# CREATES 40X40 IMAGES OF POLYNOMIALS OF MAXIMUM DEGREE
x=np.linspace(-1,1,40)
xx,yy=np.meshgrid(x,x)
       

def random_poly(deg, xx=xx, yy=yy): #function that creates polinomial of 2 variables of given degree with no cross terms
    x_params=np.random.uniform(-5.,5., deg+1)
    y_params=np.random.uniform(-5.,5., deg+1)
    
    p=np.zeros((40,40))
    
    for i in range(deg+1):
        p+=x_params[i]*xx**i + y_params[i]*yy**i 
        
    return p


# CREATES DATASET
size = 6000
data=[]

for i in range(size):
    degree=int(np.random.uniform(1,9))
    data.append(random_poly(degree))
data=np.array(data)

plt.imshow(data[1298]) # shows element 1298 of the dataset
plt.show()
```

### 2.2 Build two autoencoder architectures (e.g. similar to the ones presented in the lecture) which involve a single hidden dense layer and several hidden dense layers.

```python
# Autoencoder with a single dense hidden layer

model1=Sequential()
model1.add(Dense(800, input_dim=1600, activation='relu')) # encoder
model1.add(Dense(1600, activation='relu')) # decoder

model1.compile(loss='binary_crossentropy', optimizer='adadelta', metrics=['accuracy'])

# Arranges dataset for training

data=data.reshape((len(data), np.prod(data.shape[1:])))
x_train=data[:5900]
x_test=data[5900:]
```

```python
history1 = model1.fit(x_train, x_train, epochs=100, batch_size=256, shuffle=True, verbose=1, validation_data=(x_test, x_test))
```

```python
# Autoencoder with multiple hidden layers

model2=Sequential()
model2.add(Dense(800, input_dim=1600, activation='relu')) # encoder
model2.add(Dense(400, input_dim=1600, activation='relu')) # encoder
model2.add(Dense(200, input_dim=1600, activation='relu')) # encoder

model2.add(Dense(400, activation='relu')) # decoder
model2.add(Dense(800, activation='relu')) # decoder
model2.add(Dense(1600, activation='relu')) # decoder


model2.compile(loss='binary_crossentropy', optimizer='adadelta', metrics=['accuracy'])
```

```python
history2 = model2.fit(x_train, x_train, epochs=100, batch_size=256, shuffle=True, verbose=1, validation_data=(x_test, x_test))
```

## Ex.  3 KL – divergence

### Show by example that $D_{KL}(Q||P) \neq D_{KL}(P||Q)$ in general.

We can for example take $Q= \mathcal{G}(s,S)$ and $P= \mathcal{G}(s,S')$ where $\mathcal{G}(s,S)$ denotes the one-dimensional Gaussian with $0$ mean and variance $S$. Then

\begin{align}
D_{KL}(Q||P)=\int_\mathbb{R} G(\tilde{s},S)\ \ln\left[\frac{G(\tilde{s},S)}{G(\tilde{s},S')} \right] d\tilde{s} = \frac{1}{2} \int_\mathbb{R} \mathcal{G}(\tilde{s},S)\ \left(\ln\left[\frac{S}{S'}\right]-\frac{\tilde{s}^2}{S}+\frac{\tilde{s}^2}{S'} \right) d\tilde{s} = \frac{1}{2}\left(\ln\left[\frac{S}{S'}\right] + \frac{S}{S'} - 1\right)
\end{align}

Exchanging $\ Q \leftrightarrow P$ it becomes clear that

\begin{equation}
    D_{KL}(P||Q) = \frac{1}{2}\left(\ln\left[\frac{S'}{S}\right] + \frac{S'}{S} - 1\right) \neq \frac{1}{2}\left(\ln\left[\frac{S}{S'}\right] + \frac{S}{S'} - 1\right) = D_{KL}(P||Q)
\end{equation}

and since in principle $S \neq S'$, for generic $P$s and $Q$s $\boxed{D_{𝐾𝐿}(P||Q)\neq 𝐷_{𝐾𝐿}(Q||P)}.$
