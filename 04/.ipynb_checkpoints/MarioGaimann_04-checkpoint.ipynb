```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn.svm as svm
import sklearn.tree as tree
import sklearn.ensemble as ens
import sklearn.model_selection as skmod

import keras.models as km
import keras.layers as kl
import keras.backend as K
```

## A1 - Classifiers in sklearn

### A1a - Build classifiers with svm.svc, DecisionTreeClassifier, RandomForestClassifier and AdaBoostClassifier

```python
def con1(x,y):
    if y>=0.2*np.sin(10*x)+x:
        rv = 1
    else:
        rv = -1
    return rv

def con2(x,y):
    if (x+0.5)**2+(y-0.5)**2<= 0.1:
        rv = -2
    else:
        rv = 0
    return rv

def con3(x,y):
    if (x-0.5)**2+(y+0.5)**2<= 0.1:
        rv = 2
    else:
        rv = 0
    return rv

length=1000
X = 3.*(np.random.rand(length,2)-[0.5,0.5])
y =[con1(X[i,0],X[i,1])+con2(X[i,0],X[i,1])+con3(X[i,0],X[i,1]) for i in range(len(X))]

plt.scatter(X[:,0],X[:,1],c=y)

X_train, X_test, y_train, y_test = skmod.train_test_split(X, y)
```

```python
def plot_clf_boundaries(X, y, X_test, X_train, y_test, y_train):
    # plot classification boundaries
    plt.figure()
    plt.clf()
    plt.scatter(X[:, 0], X[:, 1], c=y, zorder=10, cmap=plt.cm.Paired,
                edgecolor='k', s=20)
    plt.scatter(X_test[:, 0], X_test[:, 1], s=80, facecolors='none',
                zorder=10, edgecolor='k')     # circle test data  

    # find range of x and y values of X input data
    x_min = X[:, 0].min()
    x_max = X[:, 0].max()
    y_min = X[:, 1].min()
    y_max = X[:, 1].max()    

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    # clf.decition_function is here also possible
    Z = clf.predict(np.c_[XX.ravel(), YY.ravel()])  
    
    Z = Z.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
                linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
```

```python
# build classifiers
clf_svm = svm.SVC(gamma='scale')
clf_tree = tree.DecisionTreeClassifier()
clf_rdfor = ens.RandomForestClassifier(n_estimators=10)
clf_adab = ens.AdaBoostClassifier(n_estimators=100)

# iterate through classifiers
classifiers = [clf_svm, clf_tree, clf_rdfor, clf_adab]
clf_labels = ["svm.svc", "decisiontree", "randomforest" ,"AdaBoost"]
scores=[]

i=0
for clf in classifiers:
    # train models and compute score with test data
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
    
    plot_clf_boundaries(X, y, X_test, X_train, y_test, y_train)
    plt.title(clf_labels[i])
    i += 1

print("Scores:\n")
print(clf_labels)
print(scores)
```

```python
# iterate through various maxdepths for trees
maxdepths = [1, 2, 5,7, 10, 100]
scores = []

for maxdepth in maxdepths:
    # train models and compute score with test data
    clf =  tree.DecisionTreeClassifier(max_depth=maxdepth)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
    
    plot_clf_boundaries(X, y, X_test, X_train, y_test, y_train)
    plt.title("decisiontree | maxdepth = %d" % maxdepth)

print("Depths:")
print(maxdepths)
print("Scores:")
print(scores)
```

A1 - Observations:

 - The Support Vector Machine with C penalty parameter for the error term (svm.svc) is not able to classify the outlying point clouds correctly and draws a straight boundary. 
 - The decision tree and random forest classifiers predict the outlying points quite precisely, only subject to small deviations from noise.
 - The AdaBoost classifier seems to build a complicated model that allows some outliers to be categorized wrongly, it features very narrow categorizations instead of bulk categories as seen for SVMs.
 
 
Comment on maximum depth of trees:

 - The higher the tree depths, the better the score gets. While a depth of 1 corresponds to a simple linear classification with a score of 0.7, deeper trees allow advanced decisions which yield a high score of about 0.96.

## A2 - XOR in Keras



```python
# define training data for XOR 
# (test data is equal to training data here)
X_train = np.asarray([[0,0],[1,0],[0,1],[1,1]])
y_train = np.asarray([[0],[1],[1],[0]])

# set up model
def create_model(X_train, y_train, validation_data, hiddenunits=16, epochs=500, weights=None):
    model = km.Sequential()
    # the relu shape seems to matter a lot here and performs better than sigmoid. Why?
    # first and only hidden layer
    model.add(kl.Dense(units=hiddenunits, activation='relu', input_shape=(2,)))
    # output layer
    model.add(kl.Dense(units=1, activation='sigmoid'))
    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    # train model
    if weights != None:
        model.set_weights(weights)
    model.fit(X_train, y_train, epochs=epochs, batch_size=1, verbose=1, validation_data = validation_data)

    return model
```

```python
# build a simple model
model = create_model(X_train, y_train, validation_data=(X_train, y_train))
model.summary()

print("Score:\n")
print(model.evaluate(X_train, y_train))
```

```python
# check model by predicting
print(model.predict(X_train).round())

```

```python
# try to utilize as few hidden neurons as possible
weights = [np.array([[1, 1],
       [1,  1 ]])]

model = create_model(X_train, y_train, (X_train, y_train), hiddenunits=2, epochs=5000, weights = weights)
model.summary()

print("Score:\n")
print(model.evaluate(X_train, y_train))

```

The minimum number of hidden units needed in this network is 2 in principle, but it heavily depends on initial conditions, which were set manually in the above run.

```python
# check model by predicting
print(model.predict(X_train).round())

print("\nNetwork Parameters:\n")
print("Hidden Layer weights:")
print(model.layers[0].get_weights()[0])
print("Hidden Layer biases:")
print( model.layers[0].get_weights()[1])
print("Output Layer weights:")
print(model.layers[1].get_weights()[0])
print("Output Layer biases:")
print(model.layers[1].get_weights()[1])
```
